```python
import random
import pandas as pd
from faker import Faker
import string
from datetime import datetime, timedelta
from random import randint, randrange
from decimal import Decimal

fake = Faker()

num_rows = 60000

file_path = 'c_s_c.xlsx'
df = pd.read_excel(file_path)

city_state_country_data = df[['city', 'state_name', 'county_name']].values.tolist()

def generate_text():
    address_remark = fake.text()
    if random.random() < 0.25:
        address_remark += " " + "".join(random.choices(string.punctuation, k=random.randint(1, 5)))
    return address_remark

data = []

# Generate data for each row and append to the data list
for i in range(num_rows):
    id = i + 1
    name = fake.name()
    age = fake.random_int(min=18, max=65)
    height = Decimal(fake.random.uniform(150, 200)).quantize(Decimal('0.00'))
    weight = Decimal(fake.random.uniform(50, 100)).quantize(Decimal('0.00'))
    salary = Decimal(fake.random.uniform(20000, 100000)).quantize(Decimal('0.00'))
    dob = fake.date_of_birth().strftime('%Y-%m-%d')
    address = fake.street_address()

    city, state, county = random.choice(city_state_country_data)

    email = fake.email()
    phone = randint(1253486907, 9999999999)
    zipcode = randint(10501, 99950)
    company = fake.company()
    job_title = fake.job()

    education_level = random.choice(['High School', 'Associate', 'Bachelor', 'Master', 'PhD'])
    degree = fake.random_element(elements=('Computer Science', 'Engineering', 'Business', 'Medicine', 'Arts', 'Science'))

    num_hobbies = fake.random_int(min=1, max=5)
    hobbies = [fake.random_element(elements=('Reading', 'Sports', 'Cooking', 'Traveling', 'Gardening', 'Painting', 'Photography', 'Gaming'))
               for _ in range(num_hobbies)]
    hobbies = ', '.join(hobbies)

    marital_status = random.choice(['Single', 'Married', 'Divorced', 'Widowed'])
    num_children = fake.random_int(min=0, max=5)

    num_interests = fake.random_int(min=1, max=3)
    interests = [fake.random_element(elements=('Music', 'Movies', 'Sports', 'Art', 'Cooking', 'Travel', 'Photography', 'Fitness', 'Technology'))
                 for _ in range(num_interests)]
    interests = ', '.join(interests)

    years_of_experience = fake.random_int(min=0, max=40)
    num_languages = fake.random_int(min=1, max=5)
    languages_spoken = [fake.language_name() for _ in range(num_languages)]
    languages_spoken = ', '.join(languages_spoken)

    linkedin_profile = f'linkedin.com/in/{name}'
    twitter_handle = f'@{name}'

    start_date_ordered = datetime.now() - timedelta(days=180)  # 6 months ago
    ordered_date = fake.date_between_dates(date_start=start_date_ordered, date_end=datetime.now())
    delivered_date = fake.date_between_dates(date_start=ordered_date, date_end=datetime.now())

    ordered_date_str = ordered_date.strftime('%Y-%m-%d')
    delivered_date_str = delivered_date.strftime('%Y-%m-%d')
    lvl = 1
    citizen = 'Y'
    remarks = fake.date_of_birth().strftime('%Y-%m-%d')
#     remark_txt = generate_text() , 'Remark_com' , remark_txt
    month_income = Decimal((salary / 12).quantize(Decimal('0.00')))

    data.append([name, id, age, height, weight, salary, dob, address, city, state, county, zipcode, email, phone, company, job_title,
                 education_level, degree, hobbies, marital_status, num_children, interests,
                 years_of_experience, languages_spoken, linkedin_profile, twitter_handle, ordered_date, delivered_date, lvl, citizen, remarks, month_income])

# Create a DataFrame from the data list
columns = ['Name', 'ID', 'Age', 'Height', 'Weight', 'Salary', 'Date_of_Birth', 'Address', 'City', 'State', 'County', 'Zipcode',
           'Email', 'Phone', 'Company', 'Job_Title', 'Education_Level', 'Degree', 'Hobbies', 'Marital_Status', 'Number_of_Children', 'Interests',
           'Years_of_Experience', 'Languages_Spoken', 'LinkedIn_Profile', 'Twitter_Handle', 'Ordered_Date', 'Delivered_Date', 'Lvl', 'Citizen', 'Remarks', 'month_income']
df = pd.DataFrame(data, columns=columns)

# Save the DataFrame to a CSV file
df.to_csv('base_data.csv', index=False, encoding='utf-8')

print('CSV file generated successfully.')
```
