```python
import pandas as pd
import random
from datetime import datetime, timedelta
import string
import re

def generate_passport_expiry():
    base_date = datetime.today()
    expiry_date = base_date + timedelta(days=random.randint(1, 365))  # Adding random days
    return expiry_date.strftime('%Y-%m-%d')


transaction_id_range = range(1001, 6001)
customer_id_range = range(101, 2601)

def ref_id_com():
    regex_pattern=r'[A-Z]{2}-[A-Z]{2}-\d{4}-[A-Z]{2}'
#     regex_pattern = r'\d{3}-[A-Z]{3}-\d{2}'
    random_letters1 = ''.join(random.choice(string.ascii_uppercase) for _ in range(2))
    random_letters2 = ''.join(random.choice(string.ascii_uppercase) for _ in range(2))
    random_letters3 = ''.join(random.choice(string.ascii_uppercase) for _ in range(2))
    random_digits = random.randint(1000, 9999)
    value = f'{random_letters1}-{random_letters2}-{random_digits}-{random_letters3}'
    if re.fullmatch(regex_pattern, value):
        return value

data = {
    'TransactionID': random.sample(transaction_id_range, 5000),
#     'CustomerID': [f'C-{cid}' for cid in random.sample(customer_id_range, 5000)],
    'CustomerID': [f'C-{cid}' for cid in random.choices(customer_id_range, k=5000)],
#     'PassportExpiryDate': [datetime.today().strftime('%Y-%m-%d') for _ in range(5000)],
    'PassportExpiryDate': [generate_passport_expiry() for _ in range(5000)],
    'Country': random.choices(['USA', 'CANADA', 'UK'], k=5000),
    'LOB': random.choices(['PrivateEquity', 'Retail', 'InvestmentBanking'], k=5000),
}

customer_passports = []
for _ in range(5000):
    passport = ref_id_com()
    customer_passports.append(passport)

data['CustomerPassport'] = customer_passports

data['Entity'] = ['EN1' if lob == 'PrivateEquity' else 'EN2' if lob == 'Retail' else 'EN3' for lob in data['LOB']]
data['LoanAmount'] = [random.randint(1000000, 2000000) if lob == 'PrivateEquity' else random.randint(1000, 2000) if lob == 'Retail' else random.randint(10000, 20000) for lob in data['LOB']]
data['InterestRate'] = [round(random.uniform(0.1, 0.15), 2) if lob == 'PrivateEquity' else round(random.uniform(0.03, 0.08), 2) if lob == 'Retail' else round(random.uniform(0.08, 0.12), 2) for lob in data['LOB']]
data['CollateralAmount'] = [amount * 5 for amount in data['LoanAmount']]

start_date = datetime(2010, 1, 1)
end_date = datetime(2023, 1, 1)
date_range = [start_date + timedelta(days=random.randint(1, (end_date - start_date).days)) for _ in range(5000)]
data['TransactionDate'] = [date.strftime('%Y-%m-%d') for date in date_range]
data['SettlementDate'] = [(date + timedelta(days=2)).strftime('%Y-%m-%d') for date in date_range]

df = pd.DataFrame(data)

df['Loanclass'] = [random.randint(1, 9) for _ in range(len(df))]

print(df.head())
```

```python
df.to_csv('bank_data_lc.csv', index=False)
```
